### Análige Geral do Perfil dos Sobreviventes

```python
#analise geral do perfil dos sobreviventes
fig = plt.figure(figsize=(22, 5))
ax = fig.add_subplot(1, 4, 1)
ax.pie(train_df.Survived.value_counts(), labels=['Mortos', 'Sobreviventes'], startangle=90, explode=[0, 0.1], shadow=True, autopct='%1.1f%%')
ax.set_title('Retrato da Tragédia')

ax = fig.add_subplot(1, 4, 2)
ax.pie(train_df[train_df.Survived==1].Sex.value_counts(), labels=train_df[train_df.Survived==1].Sex.value_counts().index, startangle=90, shadow=True, autopct='%1.1f%%')
ax.set_title('Gênero dos Sobreviventes')

ax = fig.add_subplot(1, 4, 3)
ax.pie(train_df[train_df.Survived==1].Pclass.value_counts(), labels=train_df[train_df.Survived==1].Pclass.value_counts().index, startangle=90, shadow=True, autopct='%1.1f%%')
ax.set_title('Classe Social dos Sobreviventes')

ax = fig.add_subplot(1, 4, 4)
ax.pie(train_df[train_df.Survived==1].Embarked.value_counts(), labels=train_df[train_df.Survived==1].Embarked.value_counts().index, startangle=90, shadow=True, autopct='%1.1f%%')
ax.set_title('Cidade de Embarque')

fig.subplots_adjust(top=0.8)
fig.suptitle("Análise Relativa (%)", fontsize="22");
```

**OBS:** de acordo com os dados da base de treino, podemos observar no conjunto de gráficos acima que:

* Destes que se salvaram, 68.1% eram do gênero feminino;
* Da fatia de sobreviventes, 39.8% pertenciam à 1a classe;
* Por fim, 63.8% dos sobreviventes haviam embarcado em Southampton.

### Preparando os dados para o modelo

```python
#toda a manipulacao dos dados de entrada do modelo ML será em data_X
data_X = train_df
data_X
```

```python
#tratamento dos dados da coluna Sex para que sejam representados numericamente
data_X['Gender_Code'] = LabelEncoder().fit_transform(data_X['Sex'])
data_X.drop(columns=['Sex'], inplace=True)
data_X
```

### Verificando as variáveis categóricas

```python
sns.pairplot(data_X, hue='Survived')
```

**OBS:** identifica-se Pclass, SibSp e Parch como categóricas numéricas.

### Verificando a independência das variáveis categóricas segundo método de Spearman

```python
data_X[['Pclass', 'SibSp', 'Parch']].corr(method='spearman')
```

**OBS:** nota-se uma relativa dependência entre SibSp e Parch.

### Verificando a independência das variáveis contínuas segundo método de Pearson

```python
data_X[['Age', 'Fare']].corr()
```

**OBS:** nota-se uma baixa dependência entre Age e Fare.

### Contagem de valores em branco

```python
data_X.isnull().sum()
```

### Tratamento dos valores em branco a partir do descarte

```python
data_X = data_X.dropna(subset=['Age'])
data_X.isnull().sum()
```

```python
media = data_X.Age.mean()
media_idade = {'Age': media}
data_X = data_X.fillna(media_idade)
data_X.isnull().sum()
```

### Análise do tamanho da amostra de dados

```python
data_X.info()
```

### Definindo os valores de saída

```python
data_y = data_X.Survived
data_y
```

### Definindo as features a serem empregadas

```python
features = ['Pclass', 'Gender_Code', 'Age', 'SibSp', 'Parch']
data_X = data_X[features]
data_X
```

```python
clf = LogisticRegression().fit(data_X, data_y)
clf.score(data_X, data_y)
```

```python
y_pred = clf.predict(data_X)
print(classification_report(data_y, y_pred))
```

```python
#Standardize dos valores
scaler = StandardScaler().fit(data_X)

std_X = pd.DataFrame(scaler.transform(data_X))
std_X.columns = data_X.columns
std_X
```

```python
clf = LogisticRegression().fit(std_X, data_y)
y_pred = clf.predict(std_X)
print(classification_report(data_y, y_pred))
```

### Se você tivesse que viajar no Titanic naquela época qual seria sua chance de sobreviver ao desastre?

```python
# Considerando que eu viajasse na 2a classe, genero masculino, 40 anos e desacompanhado.
d = [2, 1, 40, 0, 0]
single_X = pd.Series(data=d, index=data_X.columns)
single_X
```

```python
std_single_X = scaler.transform([single_X])
std_single_X
```

```python
y_pred = clf.predict_proba(std_single_X)
y_pred
```

**Resp:** de acordo com o modelo, eu teria apenas 19% de chance de sobreviver.

### Com o modelo em mãos o que você faria para maximizar suas chances?

```python
# Considerando que eu viajasse na 1a classe, genero masculino, 40 anos e acompanhado.
d = [1, 1, 40, 1, 1]
single_X = pd.Series(data=d, index=data_X.columns)
single_X
```

```python
std_single_X = scaler.transform([single_X])
y_pred = clf.predict_proba(std_single_X)
y_pred
```

**Resp:** para aumentar minhas chances, eu deveria viajar na 1a classe e acompanhado, por exemplo, no melhor cenário, com esposa e um filho.

**OBS:** o modelo apresenta problemas de overfit que demandaria mais tempo para análise, sobretudo, porque ele assume que se eu viajar acompanhado de N filhos, as minhas chances aumentam consideravelmente. Carece de maior análise
